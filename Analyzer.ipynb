# The following stores logs in LSM friendly Format

```python
import os, pandas as pd
def get_logs(dir='Outputs/'):
    algo=['flute','geosteiner','cockayne']
    data={'Algorithm':[],'Num_Terminals':[],'Wirelength':[],'Num Steiner Points':[],"CPU Time":[]}
    for file in os.listdir(dir):
        if file.endswith('.txt'):
            f=file.split('.')[0].split('_')
            if f[0] in algo and f[1]=='logs':
                with open(dir+file,'r') as logs:
                    for line in logs:
                        for k in data.keys():
                            if line.startswith(k):
                                s=line.split(' = ')
                                data[s[0]].append(float(s[1].split(' ')[0]))
                data['Algorithm']+=[f[0]]*int(f[2])
    return pd.DataFrame(data)     
df=get_logs()
df.to_csv('Outputs.csv')
df
```

# The following stores the logs in LSM unfriendly format

```python
import os, pandas as pd

def get_logs2(dir='Outputs/'):
    algo=['flute','geosteiner','cockayne']
    nterm=[*range(10,15)]
    rep=[10000]
    stat=['wl','nsps','ct']
    data={'terminals':[]}
    data.update({f'{a}_{s}':[] for a in algo for s in stat})
    m={'Wirelength':'wl','CPU Time':'ct','Num Steiner Points':'nsps'}
    # files=[f'{a}_logs_{r}_{t}-{t}.txt'for a in algo for r in rep for t in nterm]
    for r in rep :
        for t in nterm:
            data['terminals']+=[t]*r
            for a in algo :
                file=f'{a}_logs_{r}_{t}-{t}.txt'
                with open(dir+file,'r') as logs:
                    for line in logs:
                        if line[0] not in ['/','','\n']:
                            l=line.split(' = ')
                            if l[0] in m:data[f'{a}_{m[l[0]]}'].append(float(l[1].split(' ')[0]))
    return pd.DataFrame(data)
    # return data
df=get_logs2()
df.to_csv('Outputs2.csv')
df
```
